```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import zoom
```

```python
def convert_to_ac(grayscale_im):
    """
    Converts a grayscale image (Numpy array) to an array of attenuation coefficients

    """
    grayScaleToAttCoeff = lambda x: (x/255)*30.2
    cFunc = np.vectorize(grayScaleToAttCoeff)
    return cFunc(grayscale_im)
```

```python
path_to_slice = "validation-data\plank3\slice2-withshots.png"

#img = imread(path_to_slice)
img = Image.open(path_to_slice)

# Convert the image to grayscale
img = img.convert('L')
im_array = np.array(img)
im_array = convert_to_ac(im_array)

heatmap = plt.pcolor(im_array, cmap="coolwarm")
plt.colorbar(heatmap)
plt.title("Slice with shots (plank 3, slice 2)")
plt.show()
```

```python
path_to_slice = "validation-data\plank3\slice2.png"

#img = imread(path_to_slice)
img = Image.open(path_to_slice)

# Convert the image to grayscale
img = img.convert('L')
im_array = np.array(img)
im_array = convert_to_ac(im_array)
im_array[499, 499] = 30

heatmap = plt.pcolor(im_array, cmap="coolwarm")
plt.colorbar(heatmap)
plt.title("Slice without shots (plank 3, slice 2)")
plt.show()
```

```python
path_to_slice = "validation-data\plank3\slice2-withshots.png"

#img = imread(path_to_slice)
img = Image.open(path_to_slice)

# Convert the image to grayscale
img = img.convert('L')
im_array = np.array(img)

# resizing image
resize_factor = 75/500
resized_im = zoom(im_array, (resize_factor,resize_factor), order=1, mode = "grid-constant")
resized_im = convert_to_ac(resized_im)

heatmap = plt.pcolor(resized_im, cmap="coolwarm")
plt.colorbar(heatmap)
plt.title("Slice with shots - after resizing (plank 3, slice 2)")
plt.show()
```
